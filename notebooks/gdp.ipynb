```python

from social_data_access.social_data import SocialData
import pandas as pd
from tiny_data_warehouse import DataWarehouse
import numpy as np

sd = SocialData()

tdw = DataWarehouse()
```

```python

df2 = sd.read('world_bank.gdp_absolute')
df2 = df2[df2['Country Name'] == 'Brazil']
df = df2.to_dict(orient='records')

data = []
for i in range(1960, 2023):
    data.append([i, df[0][str(i)]])

df = pd.DataFrame(data, columns=['year', 'gdp_absolute'])
plot = df.plot(x='year', y='gdp_absolute')

type(plot)
```
