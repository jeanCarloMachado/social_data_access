```python
from social_data_access.social_data import SocialData
import pandas as pd

df = SocialData().read('world_bank.poverty')

df = df[df['Indicator Name'] == 'Poverty headcount ratio at $3.65 a day (2017 PPP) (% of population)']
df

rows = []
for i, row in df.iterrows():
    for year in range(1960, 2023):
        rows.append({
            'country': row['Country Name'],
            'year': year,
            '365_a_day_ratio_population': row[str(year)],
        })
rows

df = pd.DataFrame(rows)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
# increase the size of the plot

sns.set_theme(rc={'figure.figsize':(40,20)})
plt.figure(figsize=(2000, 300))


sns.relplot(
    data=df,
    x="year", y="365_a_day_ratio_population",
    kind='line',
    hue="country",
)

```
